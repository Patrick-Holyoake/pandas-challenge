```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file = "purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
df = pd.read_csv(file)
df.head()
```

## Player Count

* Display the total number of players


```python
filt = df.loc[:, ['SN', 'Age', 'Gender']]
filt = filt.drop_duplicates()
num_players = filt.count()
num_players
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
unique = df["Item Name"].nunique()
total_revenue = df["Price"].sum()
number_of_purchases = df["Price"].count()
average_price = total_revenue / number_of_purchases
average_price = average_price.round(2)

Purchase_Analysis_df = pd.DataFrame({
    "Unique Items": [unique],
    "Average Purchase Price": ["$" + str(average_price)],
    "Total Number of Purchases": [number_of_purchases],
    "Total Revenue": ["$" + str(total_revenue)]
})
Purchase_Analysis_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
males = df.loc[df["Gender"] == "Male", :]
purchase_per_males = males["SN"].count()
purchase_count_males = males["Item Name"].count()
total_purchase_value_males = males["Price"].sum()
average_purchase_price_males = (total_purchase_value_males / purchase_count_males).round(2)
total_unique_males = males["SN"].nunique()
purchase_total_per_person_males = (total_purchase_value_males / total_unique_males).round(2)


females = df.loc[df["Gender"] == "Female", :]
purchase_count_females = females["Item Name"].count()
total_purchase_value_females = females["Price"].sum()
average_purchase_price_females = (total_purchase_value_females / purchase_count_females).round(2)
total_unique_females = females["SN"].nunique()
purchase_total_per_person_females = (total_purchase_value_females / total_unique_females).round(2)


other = df.loc[df["Gender"] == "Other / Non-Disclosed", :]
purchase_count_other = other["Item Name"].count()
total_purchase_value_other = other["Price"].sum()
average_purchase_price_other = (total_purchase_value_other / purchase_count_other).round(2)
total_unique_other = other["SN"].nunique()
purchase_total_per_person_other = (total_purchase_value_other / total_unique_other).round(2)

Purchase_Analysis_Gender_df = pd.DataFrame({
    "Purchase Count": [purchase_count_females, purchase_count_males, purchase_count_other],
    "Average Purchase Price": ["$" + str(average_purchase_price_females), "$" + str(average_purchase_price_males), "$" + str(average_purchase_price_other)],
    "Total Purchase Value": ["$" + str(total_purchase_value_females), "$" + str(total_purchase_value_males), "$" + str(total_purchase_value_other)],
    "Avg Total Purchase per Person": ["$" + str(purchase_total_per_person_females), "$" + str(purchase_total_per_person_males), "$" + str(purchase_total_per_person_other)]
})

Purchase_Analysis_Gender_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
print(df["Age"].max())
```

```python
new_df = df[["Gender", "SN", "Age"]]
new_df
```

```python
bins = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
ages = [0,9.99,14.99,19.99,24.99,29.99,34.99,39.99,46]
new_df['Age_Bins'] = pd.cut(new_df['Age'], ages, labels= bins, include_lowest= True)
new_df.head()

totals = new_df["Age_Bins"].value_counts()
percents = totals / num_players
age_demographics = pd.Dataframe({"Total Count": totals, "Percentage of players": percents})

age_demographics = age_demographics.sort_index()
age_demographics
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
df['Age Ranges'] = pd.cut(df['Age'], ages, labels= bins, include_lowest= True)

#Calculations
age_total_purchase = df.groupby(['Age Ranges']).sum()["Price"].rename("Total Purchase Value")
age_average = df.groupby(['Age Ranges']).mean()["Price"].rename("Average Purchase Price")
age_count = df.groupby(['Age Ranges']).count()["Price"].rename("Purchase Count")
age_ave_per_person = age_total_purchase / num_players

age_data = pd.DataFrame({"Purchase Count": age_count, "Average Purchase price": age_average, "Total Purchase Value": age_total_purchase, "Average Total Purchase Per Person": age_ave_per_person})
age_data
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
user_total = df.groupby(["SN"]).sum()["Price"].rename("Total Purchase Value")
user_average = df.groupby(["SN"]).mean()["Price"].rename("Average Purchase Price")
user_count = df.groupby(["SN"]).count()["Price"].rename("Purchase Count")

user_data = pd.DataFrame({"Total Purchase Value": user_total, "Average Purchase Price": user_average, "Purchase Count": user_count})
user_sorted = user_data.sort_values("Total Purchase Value", ascending=False)
user_sorted
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
item_data = df.loc[:,["Item ID", "Item Name", "Price"]]
total_purchase = item_data.groupby(["Item ID", "Item Name"]).sum()["Price"].rename("Total Purchase Value")
ave_purchase = item_data.groupby(["Item ID", "Item Name"]).mean()["Price"].rename("Price")
count = item_data.groupby(["Item ID", "Item Name"]).count()["Price"].rename("Purchase Count")

popular_items = pd.DataFrame({"Total Purchase Value": total_purchase, "Price": ave_purchase, "Purchase Count": count})
popular_items_sorted = popular_items.sort_values("Purchase Count", ascending=False)
popular_items_sorted.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
profitable_items = popular_items.sort_values("Total Purchase Value", ascending=False)
profitable_items.head()
```
